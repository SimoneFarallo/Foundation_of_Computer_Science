# 2-For each book in the Books dataset, compute its average rating.

```python
#Prima soluzione
avg_books = pd.merge(books, books_rating, on = 'ISBN', how = 'left').groupby(['Book-Title','ISBN'], as_index = False).agg(Average_rating=('Book-Rating','mean'))
#Sono stati uniti i dataset usando come vincolo d'integrità referinziale 'ISBN', dopo aver ragruppato per 'ISBN' e titolo del libro è stata calcolata la media del rating per ogni libro.
avg_books.head(10)
```

```python
#Seconda soluzione
books_merge_rating = pd.merge(books,books_rating, on = 'ISBN')
#Unisco i due dataset attraverso il merge.
book_rating_avg = books_merge_rating.groupby('ISBN', as_index = False)[['Book-Rating']].mean()
#Raggruppiamo per 'ISBN' e calcoliamo la media del rating per ogni libro.
book_rating_avg.rename(columns = {'Book-Rating':'Average_rating'}, inplace = True)
book_rating_avg.head(10)
#A differenza del comando precedente, abbiamo ragruppato solo per 'ISBN'.
```

# 4-Merge together all rows sharing the same book title, author and publisher. We will call the resulting datset "merged books". The books that have not been merged together will not appear in merged books.

```python
#Prima soluzione
books2 = books.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index=False).count()
#Utilizziamo il groupby per raggruppare i libri che hanno stesso titolo, autore e casa editrice.
merged_books =pd.DataFrame(books2[books2['ISBN']>=2])[['Book-Title', 'Book-Author','Publisher', 'ISBN']]
merged_books
#Consideriamo solo i libri che si  presentano almeno due volte e creiamo il nuovo dataset 'merged_books'.
```

```python
#Seconda soluzione
dup_books = books[books.duplicated(subset = ['Book-Title', 'Book-Author', 'Publisher'], keep = False)].sort_values(['Book-Title', 'Book-Author', 'Publisher'])
#Abbiamo utilizzato duplicated che restituisce la serie booleana e contrassegna tutti i duplicati come veri ed abbiamo ordinato per titolo,autore e casa editrice.
merged_books = dup_books.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index = False).agg(Count =('ISBN', 'count'))
merged_books.head(5)
```

# 5 - For each book in merged books compute its average rating

```python
books_ratings_AVG = books_rating.groupby('ISBN', as_index = True).agg(Average_Rating = ('Book-Rating','mean'))
#Raggruppiamo per 'ISBN' e calcoliamo la media dei rating per ogni libro.
merged_books_ratings = pd.merge(dup_books,books_ratings_AVG, on ='ISBN', how = 'left')
merged_books_ratings.groupby('Book-Title', as_index = False).mean()
#Abbiamo unito i dataset dup_books(creato al punto 4) e books_rating_AVG tramite join left e abbiamo calcolato la media raggruppando per 'Book-Title'.
```

```python
#Media calcolata raggruppando per titolo del libro e 'ISBN'
merged_books_ratings.groupby(['Book-Title','ISBN'], as_index=False ).mean()
```

# 7 - For each book in goodbooks, compute the list of its authors. Assuming that the number of reviews with a text (column work_text_reviews_count) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the shared number of reviews with a text

```python
goodbooks['authors_splitted'] = goodbooks['authors'].str.split(',')
#Sono stati divisi gli autori usando come separatore la ','.
goodbooks['shared_number_of_reviews_with_a_text'] = goodbooks['work_text_reviews_count'] / goodbooks['authors_splitted'].apply(len)
#E' stato calcolato rapporto tra il numero di reviews e il numero di autori per ogni riga.
goodbooks_exploded = goodbooks.explode('authors_splitted')
#Col comando .explode abbiamo trasformato ogni elemento di un elenco in una riga, in modo che per ogni riga c'è solo un autore.
goodbooks_exploded['authors_splitted'] = goodbooks_exploded['authors_splitted'].str.strip()
#Il metodo strip() restituisce una copia della stringa rimuovendo sia i caratteri iniziali che quelli finali.
goodbooks_exploded.groupby(['authors_splitted'], as_index=False).sum('shared_number_of_reviews_with_a_text')[['authors_splitted', 'shared_number_of_reviews_with_a_text']]
#Abbiamo calcolato la somma di'shared_number_of_reviews_with_a_text' per ogni autore, il raggruppamento è voluto perchè un autore potrebbe aver scritto più di un libro. 
```

# 8 - For each year of publication, determine the author that has the largest value of the shared number of reviews with a text.

```python
goodbooks_sum = goodbooks_exploded.groupby(['original_publication_year', 'authors_splitted'], as_index=False).sum()
#Abbiamo raggrupato per anno di pubblicazione e autore ed abbiamo calcolato la somma del numero di reviews(in caso nello stesso anno un autore avesse pubblicato più libri).
goodbooks_max = goodbooks_sum.groupby(['original_publication_year'])['shared_number_of_reviews_with_a_text'].transform(max) == goodbooks_sum['shared_number_of_reviews_with_a_text']
#Calcoliamo il max di 'shared_number_of_reviews_with_a_text' per ogni anno di pubblicazione e lo confrontiamo con la somma calcolata al punto precedente.
best_of_year = goodbooks_sum[goodbooks_max][['original_publication_year', 'shared_number_of_reviews_with_a_text', 'authors_splitted']]
best_of_year.set_index(['original_publication_year', 'authors_splitted','shared_number_of_reviews_with_a_text']).head(10)
```

```python
#Seconda soluzione
best_of_year2 = goodbooks.loc[goodbooks.groupby('original_publication_year',as_index = True).idxmax()['shared_number_of_reviews_with_a_text']]
best_of_year2[['original_publication_year','authors', 'shared_number_of_reviews_with_a_text']].head(10)
```

# 10 - Split the users dataset according to the age. One dataset contains the users with unknown age, one with age 0-14, one with age 15-24, one with age 25-34, and so on.

```python
#Prima soluzione
users['age_group'] = pd.cut(x = users['Age'], bins = [0, 14, 24, 34, 44, 54, 64, 74, 84,94,100,1000], labels=['0-14','15-24','25-34','35-44','45-54','55-64','65-74','75-84','85-94','95-100',"over"]) 
users['age_group']
```

```python
users_NaN = users[users['age_group'].isnull()]
users_NaN.head()
#Abbiamo create un dataset contenente i soli valori nulli.
```

```python
users_15_24 = users[users['age_group']== '15-24']
users_25_34 = users[users['age_group']== '25-34']
users_35_44 = users[users['age_group']== '35-44']
users_45_54 = users[users['age_group']== '45-54']
users_55_64 = users[users['age_group']== '55-64']
users_65_74 = users[users['age_group']== '65-74']
users_75_84 = users[users['age_group']== '75-84']
users_85_94 = users[users['age_group']== '85-94']
users_95_100 = users[users['age_group']== '95-100']
#Abbiamo creato i vari dataset dividendoli per classi di età
```

```python
users_max = users[users['age_group']== 'over']
users_max
```

```python
#Seconda soluzione
intervallo = np.append(0, np.arange(14, 105, 10))
#Creiamo un array con la funzione arange che restituisce un array di valori equidistanti(ogni 10) e con l'append aggiungiamo il valore 0.
users['int_age'] = pd.cut(users['Age'], intervallo).astype(str)
#Creiamo una colonna 'int_age', che indica l'intervallo nel quale appartiene l'età, utilizzando la funzione cut.
users['int_age'].replace('NaN', 'altre età', inplace = True)
users_int_age = users.groupby('int_age')
#Ragruppiamo per intervalli di età.
dizionario = {}
for i in users_int_age.groups.keys():
    dizionario[i] = users_int_age.get_group(i)
dizionario
#Usiamo un ciclo for, per dividere gli utenti nei vari intervalli di età.
```
